### Split Validation and Test

```python
raw = '/data/qq/CSCE689/video/activitynet_music_annotation/clean_activity_net_music.v1-1.json'
with open(raw) as rfp:
    old = json.load(rfp)
```

```python
old
```

```python
trvname, vvname, tevname = [], [], []

for k,v in old['database'].items():
    if v['subset'] == 'validation':
        vvname.append(k)
    elif v['subset'] == 'training':
        trvname.append(k)
    else:
        tevname.append(k)
```

```python
len(trvname), len(vvname), len(tevname)
```

```python
import numpy as np
len(np.unique(vvname))
```

```python
vd = {'Playing bagpipes': 0,
      'Playing harmonica': 0,
      'Playing saxophone': 0,
      'Playing guitarra': 0,
      'Playing flauta': 0,
      'Playing piano': 0,
      'Playing violin': 0,
      'Playing accordion': 0,}

for vn in vvname:
    q = list(old['database'][vn]['annotations'].keys())[0]
    vd[q] += 1
```

```python
vd
```

```python
sum(vd.values())
```

```python
vd2 = {'Playing bagpipes': [],
      'Playing harmonica': [],
      'Playing saxophone': [],
      'Playing guitarra': [],
      'Playing flauta': [],
      'Playing piano': [],
      'Playing violin': [],
      'Playing accordion': [],}

for vn in vvname:
    q = list(old['database'][vn]['annotations'].keys())[0]
    vd2[q] += [vn]
```

```python
for i in vd2:
    print(len(vd2[i]))
```

```python
final_test_video_name_dict = {}
for k, v in vd2.items():
    final_test_video_name_dict[k] = v[:int(len(v)/2)]
```

```python
for i in vd2:
    print(len(final_test_video_name_dict[i]))
```

```python
final_test_video_name_list = []
for k, v in final_test_video_name_dict.items():
    final_test_video_name_list += v
```

```python
new_annotation = {}

for k, v in old['database'].items():
    if k in final_test_video_name_list:
        v['subset'] = 'testing'
    new_annotation[k] = v
    
```

```python
len(new_annotation)
```

```python
trvname, vvname, tevname = [], [], []

for k,v in new_annotation.items():
    if v['subset'] == 'validation':
        vvname.append(k)
    elif v['subset'] == 'training':
        trvname.append(k)
    else:
        tevname.append(k)

len(trvname), len(vvname), len(tevname)
```

```python
new = {'database': new_annotation, 'version': a['version']}
newpath = '/data/qq/CSCE689/video/activitynet_music_annotation/activity_net_music_w_testing.json'
with open(newpath, "w") as nfp:
    json.dump(new, nfp)
```
