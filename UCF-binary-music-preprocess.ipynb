```python
class_filename_dict = {}
for k, v in a['database'].items():
    l = v['annotations']['label']
    if l in class_filename_dict:
        class_filename_dict[l] += [k]
    else:
        class_filename_dict[l] = [k]
```

```python
qq = 0
for i in class_filename_dict:
    qq += len(class_filename_dict[i])
qq
```

```python
trvname, vvname, tevname = {}, {}, {}

for k,v in a['database'].items():
    l = v['annotations']['label']
    if v['subset'] == 'validation':
        if l in vvname:
            vvname[l] += [k]
        else:
            vvname[l] = [k]
    elif v['subset'] == 'training':
        if l in trvname:
            trvname[l] += [k]
        else:
            trvname[l] = [k]
    else:
        if l in tevname:
            tevname[l] += [k]
        else:
            tevname[l] = [k]
```

```python
temp = {}
for k in vvname:
    temp[k] = []
```

```python
for k, v in vvname.items():
    for name in v:
        if name[:-4] not in temp[k]:
            temp[k] += [name[:-4]]
```

```python
import numpy as np

test_temp = {}
val_temp = {}

for k in vvname:
    val_temp[k] = []
    test_temp[k] = []
    
for k, v in temp.items():
    split_idx = int(np.ceil(len(v)/2))
    val_temp[k] = v[:split_idx]
    test_temp[k] = v[split_idx:]
```

```python
final_vvname, final_tevname = {}, {}


for k in vvname:
    final_vvname[k] = []
    final_tevname[k] = []
    
wrong = 0
for k, v in vvname.items():
    for name in v:
        if name[:-4] in val_temp[k]:
            final_vvname[k] += [name]
        elif name[:-4] in test_temp[k]:
            final_tevname[k] += [name]
        else:
            wrong += 1
    
```
